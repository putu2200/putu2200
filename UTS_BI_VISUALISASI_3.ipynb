Membuat Pie Chart dengan Matplotlib

```python
jumlah_names = ['4098', '4374', '2155']
jumlah_counts = [2, 3, 1]
colors = ('#00FFFF', '#0000FF', '#00FF00')
explode = (0, 0, 0)
```

Pada bagian ini akan dijelaskan mengenai cara membuat diagram lingkaran.

```python
fig = plt.figure()
ax = plt.axes()
```

Menampilkan pie chart

```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(jumlah_counts, labels = jumlah_names)
```

Menambahkan nilai presentase



```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(jumlah_counts, labels = jumlah_names, autopct='%1.0f%%')
```

Menambahkan judul pada diagram



```python
fig = plt.figure(figsize=(6,6))
ax = plt.axes()
ax.pie(jumlah_counts, labels = jumlah_names, autopct='%1.0f%%')
ax.set_title('Jumlah Pembelian')
```

Kesimpulan Pie Chart-Visualisasi Data

Jumlah pembeliah diambil dari tanggal 2 adalah 50%% atau 0.5. Jumlah pembelian terbesar untuk tanggal transaksi adalah 2 yaitu sebesar 50%, sedangkan yang terkecil adalah tanggal 3 sebesar 17% saja.
